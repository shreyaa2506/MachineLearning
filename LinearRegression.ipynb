```python

# import pandas as pd
# import numpy as np
# import matplotlib.pyplot as plt
# from sklearn import linear_model


import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("homeprices.csv")
df
```

```python
plt.ylabel('price(US$)')
plt.xlabel('area(sqr ft)')
plt.scatter(df.area,df.price, color='red', marker = '*')

```

```python

reg = linear_model.LinearRegression()
reg.fit(df[['area']], df.price)
reg.predict([[3300]])
```

```python
reg.predict([[3300]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
d = pd.read_csv('areas.csv')
d.head(3)
```

```python
p=reg.predict(d)
p
```

```python
d['prices'] = p
d
```
